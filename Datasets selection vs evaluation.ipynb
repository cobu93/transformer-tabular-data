```python
import json
import pandas as pd
```

# Load results

```python
results = None

with open("eval_results.json", "r") as f:
    results = json.load(f)
```

```python
best_configs = None

with open("best_configs.json", "r") as f:
    best_configs = json.load(f)
```

```python
selected_datasets = None

with open("selected_datasets.json", "r") as f:
    selected_datasets = json.load(f)
```

# Transform results

```python
for ds_name in results:
    for agg in results[ds_name]:
        results[ds_name][agg] = results[ds_name][agg]["balanced_accuracy"]
```

```python
results_df = pd.DataFrame(results)
```

```python
best_configs_df = pd.DataFrame(best_configs)
```

```python
selected_datasets_df = pd.DataFrame(selected_datasets)
```
